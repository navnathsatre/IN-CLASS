```python
#Normalized function
def norm_fun(i):
    x= (i-i.min())/(i.max()-i.min())
    return(x)
```

```python
#normalized data frame (considering the numerical part only)
df_norm=norm_fun(univ.iloc[:,1:])
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
#normalized function
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_df_univ=scaler.fit_transform(univ.iloc[:,1:])
```

```python
#how to find optimal number of clusters
#The K mean algorithm aims to choos centroids that minimum yhe inertia, or within cluster sum of squars criterion:
```

```python
wcss=[]
for i in range(1,11):
    kmean=KMeans(n_clusters=i,random_state=0)
    kmean.fit(scaled_df_univ)
    wcss.append(kmean.inertia_)
    
plt.plot(range(1,11),wcss)
plt.title("Elbow Method")
plt.xlabel("Number of clusters")
plt.ylabel('WCSS')
plt.show()
```

```python
#Build cluster algorithm
from sklearn.cluster import KMeans
clusters_new=KMeans(4,random_state=42)
clusters_new.fit(scaled_df_univ)
```
